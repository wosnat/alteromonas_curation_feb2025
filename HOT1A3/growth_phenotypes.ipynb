```python
# Function to highlight cells based on text values
def highlight_text(s):
    if s == 'CP':
        return 'background-color: #87EBA8;'
    elif s == 'FN':
        return 'background-color: #38C477;'
    elif s == 'CN':
        return 'background-color: #F28268;'
    elif s == 'FP':
        return 'background-color: #F2543D;'
    return ''

```

```python
pdf = df[['Growth condition', 
     'Addtional compounds',
      # 'Observed normal growth', #'Simulated growth',
      # 'Simulated growth fraction',    
    'Prediction class']].pivot(columns='Growth condition', index='Addtional compounds')
# Apply the highlight function to the 'Remarks' column

pdf = pdf.style.map(highlight_text, )
pdf.to_excel('MIT1002_GF_glc_min_phenotype.xlsx')
```
